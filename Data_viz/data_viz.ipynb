```python
sns.catplot(x='CITY',y='AGE',hue='Genetic Disease',data=data1)
plt.title("AGE Vs City with Hue of  Genetic Disease")
```

```python
sns.catplot(x='CITY',y='AGE',hue='Allergies',data=data1)
plt.title("Age Vs City with Hue of Allergies")
```

```python
sns.catplot(x='CITY',y='AGE',hue=' Addiction',data=data1)
plt.title("AGE Vs City with Different Hue of Addiction")
```

```python
data_rajkot = data1[data1["CITY"] == 'Rajkot']
```

```python
ax = sns.catplot(x=' Addiction',y='AGE',data=data_rajkot)
ax.set_xticklabels(rotation=40)
plt.title("Age Vs Addiction in Rajkot City")
plt.show()
```
